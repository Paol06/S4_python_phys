## Régression polynomiale :  ``np.polyfit ()``

La fonction `polyfit` de la bibliothèque `numpy`réalise l'ajustement (_fit_) des coefficients d'un polynôme $p_n(x) = a_nx^n + ....+ a_2x^2 + a_1x + a_0$ de degré $n$ de telle sorte qu'il s'approche au mieux d'une courbe expérimentale
Les données de cette courbes, les abscisses et ordonnées $x_i, y_i$ ($i=1,..., M$), sont passées en arguments de la fonction ainsi que le degré du polynôme recherché.

Cela revient donc à minimiser la somme 
\begin{equation}
\varepsilon = \sum_{j=0}^M |p(x_j)-y_j|^2
\end{equation}
par une méthode de moindre carré.

On notera que les coefficients retournés en sortie de la fonction dans l'ordre des degrés décroissants.

La procédure d'ajustement par minimisation de l'écart peut s'appliquer à tout type de fonctions.
Néanmoins, une grande part des lois de la Physique sont décrites par des polynômes et les autres fonctions possèdent très souvent une spécificité à prendre en compte dans la procédure d'ajustement.

## ``np.polyval ()``

```python
# data creation
N = 500
Delta_x = 20
x = np.linspace (-10, 10, N)
c3, c2, c1, c0 = 1, 5, -3, -250
y = c3 * x ** 3 + c2 * x ** 2 + c1 * x + c0

c = c3, c2, c1, c0
ypolyval = np.polyval (c, x)
fig, ax = plt.subplots ()
ax.plot (x, ypolyval, linewidth = 4, linestyle = 'dashed', label = 'polyval')
ax.plot (x, y, label = 'polynôme')
ax.legend ()
plt.show ()
```
