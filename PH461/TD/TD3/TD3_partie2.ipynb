$$

$$

$$

$$

$$
N_C = 2 \left( \frac{2\pi m^*_ek_BT}{h^2} \right)^{3/2},

$$

#### Commentaire
En parallèle de l'effet Hall quantique mis en évidence par l'expérience de van Klitzing, il est apparu que la constante associée au phénomène pouvaient être mesurée avec une précision supérieure à celle des constantes fondamentales qui la définissent. [Klaus von Klitzing](https://en.wikipedia.org/wiki/Klaus_von_Klitzing) a reçu le prix Nobel de Physique en 1985 pour sa découverte de l'effet Hall quantique.

En 2018, la communauté internationale de métrologie a entériné la définition de la charge électronique à partir de la constant de Klitzing.
Le choix a été également fait de considérer certaines constantes, les plus fondamentales, comme exempte d'incertitude, c'est-à-dire connue de façon exacte (jusqu'à la mise à jour suivante). Ainsi, les incertitudes expérimentales ne sont attribuées qu'aux mesures des autres grandeurs physiques.
