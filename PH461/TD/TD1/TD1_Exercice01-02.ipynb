2. Ecrire un programme qui retourne 1 si la variable numérique $x$ est positive, -1 si elle est négative et 0 si x = 0. 

```python
x  = 2
# retourne 1 (respectivement -1) si x est positif (resp. négatif)
if x > 0 :
    print('1')
elif x < 0 :
    print('-1')
else :
    print('0')
```
