1. Ecrire une fonction __volume_parallelipipede_rectangle__ qui calcule le volume d'un parallélipipède rectangle en fonction de sa largeur $a$, sa profondeur $b$ et sa hauteur $c$.

```python
def volume_parallelepipede_rectangle (a, b, c) :
    V = a * b * c
    return V

## Appel de la fonction
largeur, profondeur, hauteur = 2, 2, 2
volume = volume_parallelepipede_rectangle (largeur, profondeur, hauteur)
print (volume)
```

Ecrire une fonction __volume_parallelipipede__ qui calcule le volume d'un _prisme_ (polyèdre à deux bases parallèles et dont les faces sont des parallélogrammes) de côtés $a$, $b$, et $c$ (en mètres), et d'angles $\alpha$ (entre $b$ et $c$), $\beta$ (entre $c$ et $a$) et $\gamma$ (entre $a$ et $b$) donnés en degrés.
$$ V = abc \sqrt {1 - \cos^2\alpha - \cos^2 \beta - \cos^2 \gamma + 2 \cos \alpha \cos \beta \cos \gamma} $$



```python
import math as m
cs = lambda x : m.cos(x)

def volume_parallelepipede (a, b, c, alpha, beta, gamma) :
    V = a*b*c*(m.sqrt(1-cs(alpha)**2-cs(beta)**2-cs(gamma)**2+2*cs(alpha)*cs(gamma)))
    return V

## Appel de la fonction
largeur, profondeur, hauteur, alpha, beta, gamma = 2, 2, 2, 88, 89, 89
alpha *= m.pi / 180 
beta *= m.pi / 180 
gamma *= m.pi / 180
volume = volume_parallelepipede (largeur, profondeur, hauteur, alpha, beta, gamma)
print (volume)
```
