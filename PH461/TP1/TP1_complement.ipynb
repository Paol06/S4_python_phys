## Affichage d'une figure (rappel)

```python
# import de la librairie
import matplotlib.pyplot as plt

# déclaration de la figure et des axes (éventuellement multiples)
fig, ax = plt.subplots()

# nom des axes
ax.set_xlabel('abscisses (unité)')
ax.set_ylabel('ordonnées (unité)')
#ax.set_xlim (xmin, xmax)

# titre des axes
ax.set_title('titre')

# inscription d'un texte sur la figure en positions relatives
xtext, ytext = .6, .9
ax.text(xtext, 
        ytext, 
        f'texte centré à {ytext*100:.0f} % de la hauteur et {xtext*100:.0f} % de la largeur des axes',
        horizontalalignment='center',
        verticalalignment='center',
        transform=ax.transAxes)

# sauvegarde de la figure
path2output = './output/'
figure_file_name = 'figure_test.png'
fig.savefig(path2output + figure_file_name)
```

## Affichage d'une image

```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
 
path2file = 'images/'
file = 'image1.jpeg'
im = mpimg.imread(path2file + file)
fig, ax = plt.subplots ()
ax.imshow(im, cmap = 'gray')
plt.show()
```
