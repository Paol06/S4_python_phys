## Programmation structurée

### Contexte 
L'analyse ionique par analyse d'ions secondaires informe sur la composition chimique d'un milieu dense (cristallisé ou amorphe).
Son principe repose sur la collecte des ions émis par le matériau mesuré sous l'effet d'un bombardement d'ions primaires (Cs$^+$, O$^+$, _etc._).

Le bombardement primaire est susceptible de modifier la morphologie de la surface qui se caractérise par sa rugosité, c'est-à-dire, le moment d'ordre 2 (l'équivalent de l'écart-type pour une fonction continue) de l'altitude topographique $h(x,y)$ :

$$
\sigma = \sqrt{ \iint_\mathrm{image} (h(x, y) - \bar{h})^2 \mathrm{d}x \mathrm{d}y }.
$$

La topographie d'un surface est mesurée au moyen d'un microscope à force atomique (AFM pour _atomic force microscope_) qui acquiert un échantillonnage de la fonction $h$ sur les $N \times M$ points d'un maillage régulier.
On considère alors la rugosité comme l'écart-type de l'image topographique :

$$
\sigma_\mathrm{rms} = \sqrt {\frac{1}{NM} \sum_{i,j=1}^{NM} (h_{i,j} - \bar{h})^2}.
$$

## Question 3 : traitement d'un lot d'images

Réutiliser les scripts précédents pour appliquer les opérations suivantes à tous les fichiers du répertoire _images_ :
1. afficher l'image et donner à la figure le nom du fichier source
2. affichage la rugosité rms en haut à droite de l'image (fonction _np.std ()_)
3. afficher sur le même graphique les profils des lignes $i = N/16$, $N/8$, $N/4$ et $N/2$ en excluant 3.125 % des bords latéraux
4. afficher en haut à droite de la figure les écarts-types de chaque profil
5. afficher en haut à gauche la moyenne des écarts-types
6. inscrire son nom de famille en bas à droite de l'image et sauvegarder la figure sous ~/PH461/TP1/output/[nom de l'image]_profils_03.png

```python
#Question 3.1

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
 
path2file = 'images/'
file = 'image1.jpeg', 'image2.jpeg', 'image3.jpeg', 'image4.jpeg', 'image5.jpeg'
im = []
for k in file:
    im.append(mpimg.imread(path2file + k))




for v in range(5):
    fig, axes = plt.subplots(2)
    shape = im[v].shape
    i = int(shape[0]/8), int(shape[0]/4), int(shape[0]/2), int(shape[0]/16),
    lim2 = int(shape[1] - shape[1]*0.03125)
    lim1 = int(shape[1]*0.03125)

    #graphe 1
    axes[0].imshow(im[v])
    axes[0].set_title(file[v])

    #erreure 1
    rms = round(im[v].std(), 3)
    axes[0].text(shape[1], -5, str(rms))

    #graphe 2
    
    im1 = []
    im2 = []
    im3 = []
    im4 = []
    im1.append(im[v][i[0], lim1:lim2])
    im2.append(im[v][i[1], lim1:lim2])
    im3.append(im[v][i[2], lim1:lim2])
    im4.append(im[v][i[3], lim1:lim2])
    x = np.linspace(lim1, lim2, len(im1[0]))

    axes[1].plot(x, im1[0], label = f"N/8 : rms = {round(np.std(im1))}")
    axes[1].plot(x, im2[0], label = f"N/4 : rms = {round(np.std(im2))}")
    axes[1].plot(x, im3[0], label = f"N/2 : rms = {round(np.std(im3))}")
    axes[1].plot(x, im4[0], label = f"N/16 : rms = {round(np.std(im4))}")
    axes[1].set_title(file[v])
    ymin, ymax = axes[1].get_ylim()
    plt.legend()
    
    #nom de famille
    axes[1].text(shape[1], -70, "Pierre-Alexis")

    #moyenne ecarts types
    moyenne = round((np.std(im1) + np.std(im2) + np.std(im3) + np.std(im4))/4, 2)
    axes[1].text(0, ymax + 5, f"rms moyen = {moyenne}")

    #sauvegarde
    plt.savefig("C:/Users/paolo/Documents/S4_python_phys/PH461/TP1/output/" + f"{file[v]}_profils_03.png")
```
